```python
fdf = df.filter(['혼인상태','가족생활에 대한 만족도','(나)가족의 수입 만족도', '(라)가족관계 만족도'])
fdf
```

```python
r, p = stats.pearsonr(fdf['혼인상태'], fdf['(라)가족관계 만족도'])
"상관 계수 : %f, p-value : %f" % (r, p)
```

```python
r, p = stats.pearsonr(fdf['가족생활에 대한 만족도'], fdf['(라)가족관계 만족도'])
"상관 계수 : %f, p-value : %f" % (r, p)
```
