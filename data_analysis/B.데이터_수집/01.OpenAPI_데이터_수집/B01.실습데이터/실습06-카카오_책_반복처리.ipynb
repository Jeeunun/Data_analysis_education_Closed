```python
import requests
import json
from pandas import DataFrame
```

```python
urlFmt = "https://dapi.kakao.com/v3/search/book?query={query}&page={page}&size={size}"
```

```python
query = "파이썬"
page = 1
size = 50
key = "19111005ad4c10663591e0a55b344ba5"
```

```python
session = requests.Session()
session.headers.update({
    "Authorization" : "KakaoAK 19111005ad4c10663591e0a55b344ba5"
})
```

```python
#다음페이지가 존재하는지 여부 -> 이 값이 False인 동안 반복(while)
isEnd = False

#반복 수행 도중 추출되는 검색결과 리스트를 결합할 빈 리스트 준비
my_list = []

#1페이지 부터 시작
page = 1

#----------이전 예제 코드 시작-------------------------
while not isEnd:
    url = urlFmt.format(query=query, page=page, size=size)
    url
    r = session.get(url)
    if r.status_code != 200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    r.encoding = "utf-8"
    data = json.loads(r.text)
    data
#-----------이전 예제 코드 끝-----------------
    searchList = data['documents']

    my_list.extend(searchList)

    isEnd = bool(data['meta']['is_end'])
    page+=1
```

```python
df = DataFrame(my_list)
df.to_excel("카카오 책검색 결과(%s).xlsx" % query)
df
```
