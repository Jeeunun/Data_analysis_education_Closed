#### Q-Q Plot

#### Ï†ïÍ∑úÏÑ± ÌåêÎã®ÏùÑ ÏúÑÌïú Kolmogorov Smironov Í≤ÄÏ†ï

```python
s, p = stats.kstest(sdf['ÏûîÏ∞®'], 'norm')
print("Kolmogorov Smirnov Test: (ÌÜµÍ≥ÑÎüâ, p-value) =", (s,p))

if p > 0.05:
    print('ÏûîÏ∞®Îäî Ï†ïÍ∑úÎ∂ÑÌè¨Î•º Îî∞Î¶ÖÎãàÎã§.(Í∑ÄÎ¨¥Í∞ÄÏÑ§ Ï±ÑÌÉù)')
else:
    print("ÏûîÏ∞®Îäî Ï†ïÍ∑úÎ∂ÑÌè¨Î•º Îã§Î•¥ÏßÄ ÏïäÏäµÎãàÎã§. (Í∑ÄÎ¨¥Í∞ÄÏÑ§ Í∏∞Í∞Å)")
```
