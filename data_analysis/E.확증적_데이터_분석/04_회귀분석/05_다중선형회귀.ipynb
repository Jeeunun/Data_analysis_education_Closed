### 데이터 가져오기

보스턴 지역의 집값(종속변수)에 영향을 주는 요인(독립변수) 분석


|필드|설명|
|---|----|
|CRIM|범죄율|
|ZN|25,000 평방피트를 초과 거주지역 비율|
|INDUS|비소매상업지역 면적 비율|
|CHAS|찰스강의 경계에 위치한 경우는 1, 아니면 0|

|RAD|방사형 고속도로까지의 거리|

|LSTAT|인구 중 하위 계층 비율|
|MEDV|집값|
|CAT.MEDV|$3000 이상 여부|

```python
df = read_excel("https://data.hossam.kr/E04/boston.xlsx")
df
```

## #02. 파이썬의 ols 객체로 분석 (맛보기)

임의의 독립변수를 선정하여 분석 수행

```python
model = ols("MEDV ~ CRIM + INDUS", data=df)
fit = model.fit()
fit.summary()
```

## #03. 모듈화 한 기능을 활용

모든 독립변수의 이름을 리스트로 생성(cls)

```python
cls = list(df.columns)
cls.remove('MEDV')
cls.remove('CAT. MEDV')
cls
```

```python
model, fit, summary, table, result, goodness, varstr = ext_ols(df, x=cls, y='MEDV')
```
