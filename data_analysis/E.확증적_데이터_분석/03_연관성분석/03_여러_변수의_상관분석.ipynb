## #03. 산점도 행렬

각 변수들을 교차로 표현한 산점도 그래프의 묶음

```python
# 그래프 초기화
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 10
plt.rcParams['figure.figsize'] = (20,20) #산점도행렬은 그래프 크기가 크다!
plt.rcParams['axes.unicode_minus'] = False
```

### 1. 기본사용

대각 원소자리에 히스토그램이 표현된다.

`pairplot()` 매서드에 `diag_kind = hist`를 적용 (기본값이므로 생략시 자동적용)

카테고리 타입은 자동으로 제외됨

```python
sb.pairplot(df2)
plt.show()
plt.close()
```

### 2. 파라미터 설정

```python
sb.pairplot(df2,
            diag_kind='kde', #대각선에 커널밀도분포 표시
            hue='vs', # 범주별 색상구분
            palette='pastel')  #pastel, bright, deep, muted, colorblind, dark
plt.show()
plt.close()

#카테고리형을 추가한 것은 범주형을 넣으려 한 것
#산점도행렬 -> 변수들을 교차시켜 연관성을 분석.
#대각선 방향으로 같음. 
```

### 3. 대각선 기준 다르게 표현하기

대각선 위 : 2차원 밀도함수 + 추세선 


대각선 아래 : 2차원 밀도함수 + 산포도

```python
g = sb.pairplot(df2,
            diag_kind='kde',
            hue='vs',
            palette='pastel') 

g.map_upper(sb.kdeplot, alpha=0.3)
g.map_lower(sb.regplot, scatter=False, truncate=False, ci=False)

plt.show()
plt.close()
```

### 상관계수 행렬

여러개의 변수를 갖는 데이터프레임에 대해서도 사용가능

```python
#범주형 변수는 제외시킨다.
df3 = df2.drop('vs', axis = 1)
corr = df3.corr(method='pearson') 
corr
```

### 산점도 행렬 시각화

```python
plt.rcParams['figure.figsize'] = (10,8)
sb.heatmap(df3.corr(method='pearson'), annot=True, cmap='summer', vmin=-1, vmax=1, linewidths=.5)
plt.show()
plt.close() 
```

### 분석결과 검정

```python
# 두개씩만 가능하다. 
print(stats.pearsonr(df3['mpg'], df3['disp']))
print(stats.pearsonr(df3['disp'], df3['hp']))
print(stats.pearsonr(df3['hp'], df3['drat']))
print(stats.pearsonr(df3['drat'], df3['wt']))
print(stats.pearsonr(df3['wt'], df3['qsec']))
print(stats.pearsonr(df3['qsec'], df3['mpg']))
```

### 반목문을 통한 일괄검정

```python
names = df3.columns
n = len(names)
pv = 0.05

data = []

for i in range(0,n):
    j = i+1 if i<n-1 else 0
    fields = names[i], 'vs' , names[j]
    s, p = stats.pearsonr(df3[names[i]], df3[names[j]])
    result = p < pv

    data.append({'fields':fields, 'statistic':s, 'pvalue':p, 'result':result})

rdf = DataFrame(data)
rdf.set_index('fields', inplace=True)
rdf
```
