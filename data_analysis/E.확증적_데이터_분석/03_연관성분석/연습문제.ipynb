## <a id='toc1_1_'></a>[필요한 패키지 설치](#toc0_)

```python
from pandas import read_excel
from scipy import stats
import numpy as np

import sys
import seaborn as sb
from matplotlib import pyplot as plt
```

## <a id='toc1_2_'></a>[문제1](#toc0_)

```python
# 데이터 불러오기
df1 = read_excel("https://data.hossam.kr/E03/baseball.xlsx")
df1
```

```python
# 상관관계 분석 - (1) 탐색적 데이터 분석
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 10
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
sb.regplot(data=df1, x='홈런수', y='평균타율')
plt.grid()
plt.show()
plt.close()
```

```python
# 상관관계 분석 - (2) numpy를 이용한 상관관계 분석
np.corrcoef(df1['홈런수'], df1['평균타율'])
```

- 결과해석 : 상관계수가 -0.074 이므로 홈런수와 평균타율은 약한 음의 상관관계를 가지고 있다.  

```python
# 상관관계 분석 - (3) 피어슨과 p-value를 이용한 상관관계 분석
r , p = stats.pearsonr(df1['홈런수'],df1['평균타율'])
'상관계수: %f, p-value: %f' %(r,p)
```

- 결과해석 : 상관계수가 -0.074이고, p-value가 0.799이므로 홈런수와 평균타율은 서로 상관관계가 없다는 귀무가설을 기각할 수 없다.

```python
# 상관관계 분석 - (4) 데이터프레임객체를 이용한 상관관계 분석
corr = df1.corr(method='pearson')
corr
```

```python
plt.rcParams['figure.figsize'] = (5,3)

corr = df1.filter(['홈런수','평균타율']).corr(method='pearson')

sb.heatmap(corr,
           annot=True,
           cmap='Blues',
           vmin=-1, vmax=1,
           linewidths=.5)

plt.show()
plt.close()
```

## <a id='toc1_3_'></a>[문제 2](#toc0_)

```python
# 데이터 불러오기
df2 = read_excel("https://data.hossam.kr/E03/product.xlsx")
df2
```

```python
# 상관관계 분석 - (1) 탐색적데이터 분석
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 10
plt.rcParams['figure.figsize'] = (10,7)
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
sb.regplot(data=df2, x='x', y='y')
plt.grid()
plt.show()
plt.close()
```

```python
# 상관관계분석 - (2) numpy를 이용한 상관관계분석
np.corrcoef(df2['x'],df2['y'])
```

- 결과해석 : 두 변수의 상관계수는 0.888 이므로 강한 양의 상관관계를 갖고 있다.

```python
# 상관관계 분석 - (3) 피어슨, p-value를 이용한 상관관계분석
r, p = stats.pearsonr(df2['x'], df2['y'])
"상관계수 : %f, p-value : %f" %(r,p)
```

- 결과해석 : 두 변수의 상관계수는 0.888이고, p-value는 0.000599이므로 0.05보다 작기때문에 귀무가설을 기각할때의 위험도가 낮으므로 대립가설을 채택. 즉, x와 y는 유의한 상관관계가 있다고 볼 수 있다.

```python
# 상관관계 분석 - (4) 데이터프레임 객체를 이용한 상관관계분석
corr = df2.corr(method='pearson')
corr
```

```python
plt.rcParams['figure.figsize'] = (5,3)

corr = df2.filter(['x','y']).corr(method='pearson')

sb.heatmap(corr,
           annot=True,
           cmap='Greens',
           vmin=-1, vmax=1,
           linewidths=.5)

plt.show()
plt.close()
```

## <a id='toc1_4_'></a>[문제3](#toc0_)

```python
# 데이터 불러오기
df3 = read_excel("https://data.hossam.kr/E03/study_time.xlsx")
df3
```

```python
# 상관관계분석 - (1) 탐색적 데이터 분석
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 10
plt.rcParams['figure.figsize'] = (10,7)
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
sb.regplot(data=df3, x='공부시간', y='성적')
plt.grid()
plt.show()
plt.close()
```

```python
# 상관관계분석 - (2) numpy를 이용한 상관관계분석
np.corrcoef(df3['공부시간'], df3['성적'])[0,1]
```

- 결과해석 : 두 변수의 상관계수는 0.684 이므로 약한 양의 상관관계를 갖는다.

```python
# 상관관계 분석 - (3) 피어스만, p-value를 이용한 상관관계분석
r, p = stats.pearsonr(df3['공부시간'], df3['성적'])
'상관계수: %f, p-value: %f' %(r,p)
```

- 결과해석 : 두 변수의 상관계수는 0.684 이고, p-value는 0.029이다. 유의수준 0.05보다 작으므로 귀무가설을 기각시킬 위험도가 낮으므로 대립가설을 채택. 즉, 공부시간과 성적은 서로 유의한 관계에 있다고 볼 수 있다.

```python
# 상관관계 분석 - (4) 데이터프레임 객체를 이용한 상관관계분석
corr = df3.corr(method='pearson')
corr
```

```python
plt.rcParams['figure.figsize'] = (5,3)

corr = df3.filter(['공부시간','성적']).corr(method='pearson')

sb.heatmap(corr,
           annot=True,
           cmap='Pastel1',
           vmin=-1, vmax=1,
           linewidths=.5)

plt.show()
plt.close()
```
