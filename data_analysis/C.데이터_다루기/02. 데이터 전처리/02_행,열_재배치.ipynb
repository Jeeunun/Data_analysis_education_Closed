### #02. 행과 열의 재배치

##### 1. 열순서 변경
###### 지정된 순서대로 열이 재정렬된 결과가 반환된다.
###### 원본은 변화 없음. 결과가 적용된 복사본이 생성된다.
###### 재배치 되지 못한 열은 제외된다.

```python
df1 = df.reindex(columns=['국어','수학','과학','영어'])
df1
```

##### 2. 행순서 변경
###### 지정된 순서대로 열이 재정렬된 결과가 반환된다.
###### 원본은 변화 없음. 결과가 적용된 복사본이 생성된다.
###### 재배치 되지 못한 열은 제외된다.

```python
df2 = df.reindex(index=['석영','호영','민철','수현'])
df2
```

##### 3. 행, 열의 이름 변경
###### {'기존이름' : '새 이름'}
###### columns와 index 중 변경을 원하는 하나만 설정 가능
###### 원본은 변화없음, 결과가 적용된 복사본이 생성된다.
###### 이름이 변경되지 않은 항목이 결과에서 제외되지는 않는다. 
###### 원본에 즉시 반영해야 할 경우, inplace = True 파라미터 적용 

```python
df3 = df.rename(
    columns= {'국어' : 'kor', '영어': 'eng', '수학': 'math', '과학' : 'sci'},
    index= {'석영': 'suk', '호영':'ho', '민철':'min', '수현':'su'}
)
df3
```

### #05. 행 혹은 열 추가 삭제

### 1. 행 추가 (원본 자체가 수정된다.)

#### 리스트로 추가하기
- 추가될 행의 인덱스 이름을 지정해야 한다.
- 리스트로 추가할 경우 dataFrame의 칼럼 순서에 맞게 지정해야 한다.
- 누락되는 값이 있거나 값의 수가 초과할 경우 에러

```python
df.loc['정호'] = [2, '남자', 90, 80, 70, 60]
df
```

#### 딕셔너리로 추가하기
- 칼럼 순서는 상관없다. (key, value 값이 존재해서)
- 누락되는 값이 있거나 값의 수가 초과할 경우 에러

```python
df.loc['민정'] = {'성별':'여자', '학년':4, '국어':81, '영어':72, '수학':84, '과학':90}
df
```

#### 시리즈 객체로 추가하기
- 시리즈 객체의 index가 지정되지 않은 경우 값을 추가할 수 없다.

```python
s = Series([2,'남자',90,80,70,60], index=['학년','성별','국어','영어','수학','과학'])
df.loc['형석'] = s
df
```

#### 빈값이 있는 경우
- Series 객체 형태만 가능

```python
s = Series([2,'남자',93,71,68], index=['학년','성별','국어','수학','과학'])
df.loc['석호'] = s
df
```

### 기존의 행 복사하기
