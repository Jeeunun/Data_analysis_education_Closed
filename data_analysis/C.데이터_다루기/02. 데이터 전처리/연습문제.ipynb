```python
# 인자의 수준을 나타내는 변수를 class로, 실습값을 time으로 하여 데이터를 구성하여라
# 풀이 1. 리스트를 원소로 갖는 딕셔너리 활용 (딕셔너리의 key = column명)
data = {
    "class" : ["A1", "A1", "A1", "A1", "A1",
               "A2", "A2", "A2", "A2", "A2", "A2",
               "A3", "A3", "A3", "A3", "A3", "A3", "A3",
               "A4", "A4", "A4", "A4", "A4", "A4", "A4"],
    "time" : [62,60,63,59,61,
              62,67,71,64,65,66,
              68,66,71,67,68,69,70,
              58,62,60,59,61,63,64
              ]
}

df = DataFrame(data)

# 풀이 2. 딕셔너리를 원소로 갖는 리스트 활용
data = [
    {"class": "A1", "time":62},
    {"class": "A1", "time":60},
    {"class": "A1", "time":63},
    {"class": "A1", "time":59},
    {"class": "A1", "time":61},
    {"class": "A2", "time":62},
    {"class": "A2", "time":67},
    {"class": "A2", "time":71},
    {"class": "A2", "time":64},
    {"class": "A2", "time":65},
    {"class": "A2", "time":66},
    {"class": "A3", "time":68},
    {"class": "A3", "time":66},
    {"class": "A3", "time":71},
    {"class": "A3", "time":67},
    {"class": "A3", "time":68},
    {"class": "A3", "time":69},
    {"class": "A3", "time":70},
    {"class": "A4", "time":58},
    {"class": "A4", "time":62},
    {"class": "A4", "time":60},
    {"class": "A4", "time":59},    
    {"class": "A4", "time":61},
    {"class": "A4", "time":63},
    {"class": "A4", "time":64}          
]

df = DataFrame(data)
df
```

```python
# 추가내용
# 생성된 데이터 프레임에 대한 각 변수의 데이터 타입 확인
df.dtypes
```

```python
# (1) 카테로기형으로 변경해야 하는 경우
# 왜 바꿀까? 문자열을 카테고리형으로 바꾸면 메모리 저장 효율적 = 빨리 계산 가능. 
# - 1) 데이터프레임의 여러 컬럼을 일괄 변경하여 새로운 데이터프레임 생성
df_1 =df.astype({'class':'category'})
df_1.dtypes

# - 2) 특정컬럼의 타입만 변경 후 원본에 바로 적용
df['class'] = df['class'].astype('category')
df.types

# (2) 서열척도로 변경해야 하는 경우
from pandas import CategoricalDtype
df['class'] = df['class']. astype(CategoricalDtype(categories=['A1','A2','A3','A4'], ordered=True))
df.dtypes
```

### 문제3

#### 내 풀이

```python
# step1. 샘플 데이터 가져오기
df3 = read_excel("https://data.hossam.kr/test/q3.xlsx")
df3
```

```python
#step2. 구성=공학 행 지우기
idx = df3[df3['구성']=='공학'].index
df3.drop(idx, inplace=True)
df3
```

```python
# step3. 도시와 농촌별에 따른 남여 평균 흡연율 구하기
pivot_table(df3,
            index='지역',
            columns='구성',
            values='흡연율',
            aggfunc='mean')
```
