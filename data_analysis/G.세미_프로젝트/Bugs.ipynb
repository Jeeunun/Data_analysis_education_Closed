```python
import datetime as dt
import pandas as pd
import time
import bs4
import requests
session=requests.Session()
session.headers.update({
    'Refer' : "",
    "user-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
})
```

```python
def transtime(x):
    return "".join(str(x).split(" ")[0].split("-"))
```

```python
days=list(map(transtime,pd.date_range(pd.to_datetime("20230102"),dt.datetime.now(),freq="6D")))
Bugs={}
Artists=[]
collection=['title','artist']
```

```python
artist_url={}
for day in days:
    bugs=session.get(f'https://music.bugs.co.kr/chart/track/week/kpop?chartdate={day}')
    time.sleep(1)
    soup=bs4.BeautifulSoup(bugs.text)
    ranking_data=[]
    for info in soup.select(".list.trackList.byChart")[0].select("tbody>tr"):
        insert_data={}
        insert_data["ranking"]=str(info.select(".ranking")[0]).split("strong")[1][1:-2]
        for col in collection:
            insert_data[col]=info.select("."+col+">a")[0].text.strip()
        # 아티스트 데이터 수집하기 위한 url 저장
        artist_url[str(info.select(".artist>a")[0].attrs["title"]).strip()]=info.select(".artist>a")[0].attrs["href"]
        # ranking 데이터 수집 
        ranking_data.append(insert_data)
    Bugs[day]=ranking_data
```

#### 아티스트 데이터 수집

```python
for artist in artist_url:
    insert={"name":artist}
    art=session.get(artist_url[artist])
    time.sleep(2)
    soup=bs4.BeautifulSoup(art.text)
    for data in soup.select(".info>tbody>tr"):
        if data.select("th")[0].text.strip()=="유형":
            insert["type"]=data.select("td")[0].text.strip()
            if data.select("td")[0].text.strip().find("그룹")!=-1:
                continue
            else:
                pass
        if data.select("th")[0].text.strip()=="멤버":
            insert["member"]=",".join(map(lambda x : x.strip(),data.select("td")[0].text.split(",")))
    Artists.append(insert)
```
