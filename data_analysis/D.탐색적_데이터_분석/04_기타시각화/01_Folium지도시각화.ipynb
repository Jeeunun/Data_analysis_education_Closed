### #02. 지도 표현하기
#### 1) 지도 객체 생성 (folium.Map(location=[위도, 경도], zoomstart=?))
: 지도의 중심이 되는 위도와 경도를 설정

```python
# zoom_start : 배율 1~22
# 주소 복사 -> 3d...(=위도) !4d....(=경도) 
map_osm = folium.Map(location=[37.5004444, 127.0313861], zoom_start=17)
map_osm
```

#### 2) 지도객체에 마커 추가 folium.Marker([위도, 경도], popup=? ,Icon=?)
: 일반마커 <br>
: 아이콘 색상값 종류 <br>
> 'lightgreen', 'darkgreen', 'darkblue', 'cadetblue', 'orange', 'lightred', 'darkred', 'green', 'blue', 'black', 'lightblue', 'white', 'lightgray', 'red', 'pink', 'beige', 'gray', 'purple', 'darkpurple'

```python
# 새로운 지도 객체 생성
map_osm1 = folium.Map(location=[37.5004444, 127.0313861], zoom_start=17)

# 마커 객체 생성
# -> 마커가 표시될 위치 [위도, 경도], 클릭시 표시될 메세지 및 아이콘 색상
marker1 = folium.Marker([37.5004444, 127.0313861], popup='메가스터디IT아카데미강남',
                        icon=folium.Icon(color='darkred'))
# -> 마커 객체를 지도에 추가함
marker1.add_to(map_osm1) 
#-> 지도 표시하기
map_osm1
```

#### 3) 사용자지정 아이콘, HTML 팝업

```python
# 새로운 지도 객체 생성
map_osm2 = folium.Map(location=[37.5004444, 127.0313861], zoom_start=17)

# HTML을 사용한 팝업
popup_html = folium.Popup("<div style='white-space: nowrap'><h3>메가스터디IT아카데미 강남</h3><img src='https://img.megastudyacademy.co.kr/campus/intra/main/B0001/m/bg_gm_01.jpg'>")

# 사용자 지정 아이콘 이미지 사용
# -> 온라인 상의 url, 내 컴퓨터 상의 상대,절대경로 모두 가능함
icon_img = folium.features.CustomIcon('https://data.hossam.kr/D04/ho.png', icon_size=(50,50))

# 마커 객체 생성
marker2 = folium.Marker([37.5004444, 127.0313861], 
                        popup = popup_html,
                        icon = icon_img)

# 마커 객체를 지도에 추가함
marker2.add_to(map_osm2)

# 웹 페이지 파일이 저장될 폴더 생성
if not path.exists('output'):
    mkdir('output')

# 파일이 저장될 위치
map_osm2.save('output/map_osm2.html')

# 지도에 표시하기
map_osm2
```

#### 4) 원형 마커(범위지정)

```python
# 새로운 지도 객체 생성
map_osm3 = folium.Map(location=[37.5004444, 127.0313861], zoom_start=17)

# 마커 객체 생성
marker1 = folium.Marker([37.5004444, 127.0313861], icon=folium.Icon(color='orange'))

# 원형마커
marker2 = folium.CircleMarker([37.5004444, 127.0313861],
                              radius=100, #범위
                              color='#3186cc', #선 색상
                              fill_color='#3186cc' #면 색상
                              )

# 원형 마커 위에 아이콘 마커 올리기
marker2.add_to(map_osm3)
marker1.add_to(map_osm3)

# 마커 객체 생성
marker3 = folium.Marker([37.5005454, 127.0314871],  
                        icon=folium.Icon(color='red'))

map_osm3
```

### #04. 예제. 서울의 고등학교 분포 확인하기

#### 1. 데이터 준비하기
: 전국 초,중,고 학교 위치 엑셀 파일을 원격으로 내려받아 데이터프레임으로 변환(다소 시간이 소요)

```python
xlsx = read_excel("https://data.hossam.kr/D04/school.xlsx")
xlsx
```

#### 2. 데이터 전처리
: 1> 사용할 필드만 추출

```python
df = xlsx.filter(['학교명','학교급구분','소재지도로명주소','위도','경도'])
df
```

: 2> 서울시의 고등학교만 추출
> Like연산 : 컬럼이름.str.contains('검색어')

```python
df2 = df.query("학교급구분 =='고등학교' and 소재지도로명주소.str.contains('서울')" )
df2
```

#### 3. 데이터 시각화

```python
# zoom_start : 배율 1~22
map_osm = folium.Map(location=[37.566651, 126.978428], zoom_start=12)

html = "<font color='green' style='white-space: nowrap'><b>%s</b></font>"

# 데이터프레임의 행 수 만큼 반복하면서 마커 생성
for i in df2.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = df2.loc[i,'학교명']
    lat = df2.loc[i, '위도']
    ing = df2.loc[i, '경도']

    # 추출한 정보를 지도에 표시
    popup_html = folium.Popup(html % name, parse_html=False)
    marker = folium.Marker([lat, ing], popup=popup_html)
    marker.add_to(map_osm)

map_osm
```
